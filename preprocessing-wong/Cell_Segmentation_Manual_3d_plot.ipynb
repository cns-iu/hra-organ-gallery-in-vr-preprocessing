```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("CD8_stats_Position.csv")
```

```python
df.head()
```

```python
df[["Position X", "Position Y", "Position Z"]] = df[["Position X", "Position Y", "Position Z"]] * 1e-6
```

```python
fig = px.scatter_3d(
    df,
    x="Position X",
    y="Position Y",
    z="Position Z",
    hover_name="ID",
    title="3D Position Visualization",
    labels={"PositionX": "X (µm)", "PositionY": "Y (µm)", "PositionZ": "Z (µm)"},
)

fig.update_layout(
    width=1000,
    height=800
)

fig.show()
```
