## Code to preprocess Chenchen Zhu's data to form aggregates of all cell types that are not top 5 occuring 

For Cell Layers:

```python
import pandas as pd
import os

# Define the data folder and file paths
data_folder = "data"
# files = ["layer_16.csv", "rna_16.csv"]
files = ["layer_16.csv"]

# Process each file
for file in files:
    file_path = os.path.join(data_folder, file)
    
    # Load CSV
    df = pd.read_csv(file_path)
    
    # Check if 'Cell Type' column exists
    if 'Cell Type' not in df.columns:
        print(f"Skipping {file}: 'Cell Type' column not found.")
        continue
    
    # Find top 5 cell types by frequency
    top5 = df['Cell Type'].value_counts().nlargest(5).index
    
    # Replace other values with 'Other'
    df['Cell Type'] = df['Cell Type'].apply(lambda x: x if x in top5 else 'Other')
    
    # Define new output filename
    base_name, ext = os.path.splitext(file)
    new_file = f"{base_name}_cleaned{ext}"
    new_path = os.path.join(data_folder, new_file)
    
    # Save to new file
    df.to_csv(new_path, index=False)
    
    print(f"Processed {file}: saved cleaned data as {new_file}")

print("All files processed successfully and saved as new copies.")
```
