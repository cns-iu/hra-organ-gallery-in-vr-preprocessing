CSV crosswalk splitter

```python
import os
import pandas as pd

def generate_modified_csvs(input_csv_path):
    # Ensure file exists
    if not os.path.isfile(input_csv_path):
        raise FileNotFoundError(f"File not found: {input_csv_path}")
    
    # Extract base file name (without extension)
    base_name = os.path.splitext(os.path.basename(input_csv_path))[0]
    
    # Read full CSV
    df = pd.read_csv(input_csv_path)
    
    # Expected level columns and output suffix
    level_columns = {
        "Level Three Cell Type": "Level3",
        "Level Two Cell Type": "Level2",
        "Level One Cell Type": "Level1"
    }
    
    # Validate required columns
    required_columns = ["x", "y"] + list(level_columns.keys())
    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Missing required column: {col}")
    
    # Create output directory
    output_dir = "Data"
    os.makedirs(output_dir, exist_ok=True)
    
    # Generate one file per level
    for level_name, suffix in level_columns.items():
        output_filename = f"{base_name}_{suffix}.csv"
        output_path = os.path.join(output_dir, output_filename)
        
        # Rename selected level column to "Cell Type"
        temp_df = df[["x", "y", level_name]].copy()
        temp_df = temp_df.rename(columns={level_name: "Cell Type"})
        
        # Sort alphabetically by Cell Type
        new_df = temp_df.sort_values(by="Cell Type", kind="mergesort")
        
        # Save to CSV
        new_df.to_csv(output_path, index=False)
        
        print(f"Created: {output_path}")
```
